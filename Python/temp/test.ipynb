```python
import pandas as pd
```

```python
df = pd.read_csv('data/부산광역시_부산진구_가로수현황_20201123.csv', encoding='cp949')
df = pd.DataFrame(df)
df.head(2)
```

```python
df = df[['위치명', '위도', '경도']]
df.head()
```

```python
df['위도'] = pd.to_numeric(df['위도'])
df['경도'] = pd.to_numeric(df['경도'])

# df.loc[31]
```

```python
import folium

#지도의 중심을 지정하기 위해 위도와 경도의 평균 구하기
lat = df['위도'].mean()
# lat = ex['위도'].mean()
long = df['경도'].mean()

#지도 띄우기
m = folium.Map([lat, long], zoom_start=9)
m
```

```python
#지도 새로 띄우기
m = folium.Map([lat, long], zoom_start=15)

for i in df.index:
    sub_lat = df.loc[i, '위도']
    sub_long = df.loc[i, '경도']
    title = df.loc[i, '위치명']

    color = 'blue'
    # if df.loc[i, '구분'] == '소매':
    #     color = "red"

    #지도에 동그라미로 데이터 찍기
    folium.CircleMarker([sub_lat, sub_long], color=color, fill_color='orange', radius=15, tooltip=title).add_to(m)

#한글이 안나오는 오류로 html로 trouble shooting
m.save('map.html')
m
```
