### 구별로 변경
---

```python
import pandas as pd
```

```python
df = pd.read_csv('result/부산_동별_방제취약지수_최종_euckr.csv', encoding='euc-kr')
df.head()
```

```python
df.columns
```

```python
df2 = df[['구', '총인구', '세대수', '공원개수', '가로수개수']]
df3 = df[['구', '평균기온', '최고기온', '최저기온', '강수량', '습도', '주거지부', '공원부', '수변부', '해충지수', '해충취약지수', '해충취약지수2']]
df3.head()
```

```python
df22 = df2.groupby(df['구'], as_index=True).sum()
df22.head()
```

```python
df33 = df3.groupby(df['구'], as_index=True).mean()
df33.head()
```

```python
dfdf = pd.merge(df22, df33, left_on='구', right_on='구')
dfdf.head()
```

```python
dfdf.to_csv('result/부산_구별_해충취약지수_최종_euckr.csv', encoding='euc-kr')
```

```python
po = pd.read_csv('data/구별 포충기 설치 대수.csv')
po.head()
```

```python
po.columns = ['구', '포충기대수']
```

```python
dfdf2 = pd.merge(dfdf, po, left_on='구', right_on='구')
dfdf2.head()
```

```python
dfdf2.to_csv('result/부산_구별_해충취약지수_최종_euckr.csv', encoding='euc-kr', index=False)
```
