```python
# 중복 동 묶기
# 일단 앞에 두글자 따와서
def dong_concat(df):
    dongs = list()
    for dong in df['동']:
        if(dong[:1] == '좌' or dong[:1] == '우' or dong[:1] == '중' or dong[:1] == '서'):
            if(dong[:2] == '서대'):
                dongs.append('서대신동')
            elif(dong[:2] == '중앙'):
                dongs.append('중앙동')
            elif(dong[:2] == '우암'):
                dongs.append('우암동')
            elif(dong[:2] == '좌천'):
                dongs.append('좌천동')
            else:
                dongs.append(dong[:1]+'동')
        else:
            if(dong[:3] == '금사회'):
                dongs.append('금사동')
            elif(dong[:2] == '기장'):
                dongs.append('기장읍')
            elif(dong[:2] == '가덕'):
                dongs.append("가덕도동")
            elif(dong[:2] == '청룡'):
                dongs.append("청룡노포동")
            elif(dong[:2] == '남부'):
                dongs.append('남부민동')
            elif(dong[:2] == '철마'):
                dongs.append('철마면')
            elif(dong[:2] == '정관'):
                dongs.append('정관읍')
            elif(dong[:2] == '장안'):
                dongs.append('장안읍')
            elif(dong[:2] == '선두'):
                dongs.append('선두구동')
            elif(dong[:2] == '일광'):
                dongs.append('일광면')
            elif(dong[:2] == '남부'):
                dongs.append('남부민동')
            elif(dong[:2] == '동대'):
                dongs.append('동대신동')
            else:
                dongs.append(dong[:2]+'동')
    df['통합동'] = dongs
    return df
```

```python
dong_concat(pop_g)
```

```python
gu = pop_g[['구', '통합동']]
gu.head()
```
