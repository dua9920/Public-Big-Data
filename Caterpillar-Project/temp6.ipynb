```python
import pandas as pd
import os
```

```python
# 경로 변경
# os.chdir('result/6월')
os.chdir('../8월')
```

```python
# 파일 리스트 가져오기
pest = list(os.listdir(os.getcwd()))
print(len(pest), pest[0])
```

```python
# 1일 기준으로 동별 행 concat 하기
merged_df = pd.DataFrame()
for i in pest:
    df = pd.read_csv(i)
    day = df.loc[0]
    merged_df = merged_df.append(day)
merged_df.reset_index(drop=True, inplace=True)
merged_df.to_csv('../9월_모기지수_1일기준.csv', encoding='utf-8', index=False)
```

```python
# 저장한 파일 가져오기
result = pd.read_csv('../9월_모기지수_1일기준.csv')
result.head()
```

```python
# 중복 동 묶기
# 일단 앞에 두글자 따와서
dongs = list()
for dong in result['동이름']:
    if(dong[:1] == '좌' or dong[:1] == '우' or dong[:1] == '중' or dong[:1] == '서'):
        if(dong[:2] == '서대'):
            dongs.append('서대신동')
        elif(dong[:2] == '중앙'):
            dongs.append('중앙동')
        elif(dong[:2] == '우암'):
            dongs.append('우암동')
        elif(dong[:2] == '좌천'):
            dongs.append('좌천동')
        else:
            dongs.append(dong[:1]+'동')
    else:
        if(dong[:3] == '금사회'):
            dongs.append('금사동')
        elif(dong[:2] == '기장'):
            dongs.append('기장읍')
        elif(dong[:2] == '가덕'):
            dongs.append("가덕도동")
        elif(dong[:2] == '청룡'):
            dongs.append("청룡노포동")
        elif(dong[:2] == '남부'):
            dongs.append('남부민동')
        elif(dong[:2] == '철마'):
            dongs.append('철마면')
        elif(dong[:2] == '정관'):
            dongs.append('정관읍')
        elif(dong[:2] == '장안'):
            dongs.append('장안읍')
        elif(dong[:2] == '선두'):
            dongs.append('선두구동')
        elif(dong[:2] == '일광'):
            dongs.append('일광면')
        elif(dong[:2] == '남부'):
            dongs.append('남부민동')
        elif(dong[:2] == '동대'):
            dongs.append('동대신동')
        else:
            dongs.append(dong[:2]+'동')
# print(dongs)
result['통합동'] = dongs
result.head()
```

```python
# 그룹핑
result_g = result.groupby(result['통합동'], as_index=False).mean()
result_g
```

```python
# 최종 파일 저장
result_g.to_csv('../9월_모기지수_1일기준_최종.csv', index=False)
```
