### 포충기 개수, 가로등
---

```python
# 포충기 파일 (구별)
po = pd.read_csv('data/구별 포충기 설치 대수.csv')
po.head()
```

```python
# 가로등 설치 수 (구별)
light = pd.read_csv('data/부산광역시_가로등(광원별) 설치 현황_20200101.csv', encoding='euc-kr')
light.head()
```

```python
# 가로등 총 합 구하기
light_sum = list()
for i in range(len(light)):
    tmp = light.iloc[i][1:].sum()
    light_sum.append(tmp)
light_sum
```

```python
# 열 추가하기 1
light_res =pd.DataFrame(columns={'구분'})
light_res['구분'] = light['구분']
light_res = light_res.drop(light_res.index[16:])
light_res
```

```python
# 열 추가하기 2
light_res['가로등개수'] = light_sum[:16]
light_res.head()
```

```python
light_res.to_csv('result/부산_구별_가로등수.csv', encoding='utf-8', index=False)
```
