### 부산광역시 해충지수 평균 + 방제취약지수
----

```python
import pandas as pd
```

```python
df = pd.read_csv('result/부산_8월_동별_최종.csv')
df.head()
```

```python
df['해충지수'] = round(df['수변부'] + df['공원부'] + df['주거지부'] /3, 4)
df.head()
```

```python
df['방제취약지수'] = round((df['해충지수'] * 0.5) + (df['습도'] * 0.2) + (df['가로수개수'] * 0.5), 4)
df.head()
```

```python
df.describe()
```

```python
df = df[['통합동', '구', '평균기온', '최고기온', '최저기온', '강수량', '습도', '공원개수', '가로수개수', '공간유형', '주거지부', '공원부', '수변부', '해충지수', '방제취약지수']]
df.head(2)
```

```python
# 총인구 세대수 불러오기 + 병합하기
pop_df = pd.read_csv('result/총인구/동별_총인구_세대수_결과.csv')
pop_df.head()
```

```python
pop_df = pop_df[['통합동', '총인구', '세대수']]
pop_df.head()
```

```python
df = pd.read_csv('result/부산_동별_방제취약지수_최종.csv')
df.head(2)
```

```python
df = pd.merge(df, pop_df)
df.head(2)
```

```python
df = df[['통합동', '구', '총인구', '세대수', '평균기온', '최고기온', '최저기온', '강수량', '습도', '공원개수', '가로수개수', '공간유형', '주거지부', '공원부', '수변부', '해충지수', '방제취약지수']]
df.head(2)
```

```python
df.to_csv('result/부산_동별_방제취약지수_최종.csv', index=False)
```

```python
df.to_csv('result/부산_동별_방제취약지수_최종_euckr.csv', encoding='euc-kr', index=False)
```
