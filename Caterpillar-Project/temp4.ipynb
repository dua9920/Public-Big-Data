```python
# 반복문으로 통일
for i in range(0, len(weather_list)-2, 5):
    # 파일 읽어오기
    temp = pd.read_csv(f'{weather_list[i]}')
    preci = pd.read_csv(f'{weather_list[i+1]}')
    hum = pd.read_csv(f'{weather_list[i+2]}')
    high = pd.read_csv(f'{weather_list[i+3]}')
    low = pd.read_csv(f'{weather_list[i+4]}')

    if(weather_list[i].find('동', 2) == -1):
        if(weather_list[i].find('읍') == -1):
            index = weather_list[i].find('면', 2)
            name = weather_list[i][:index] + '면'
        else:
            index = weather_list[i].find('읍', 2)
            name = weather_list[i][:index] + '읍'
    else:
        index = weather_list[i].find('동', 2)
        name = weather_list[i][:index] + '동'

    # 열 이름 바꾸기
    temp.columns = ['일', '시간', 'forecast', '평균기온']
    preci.columns = ['일', '시간', 'forecast', '강수량']
    hum.columns = ['일', '시간', '습도']
    high.columns = ['일', '시간', 'forecast', '최고기온']
    low.columns = ['일', '시간', 'forecast', '최저기온']

    # 필요없는 열 제거
    temp = temp[['일', '평균기온']]
    preci = preci[['일', '강수량']]
    hum = hum[['일', '습도']]
    high = high[['일', '최고기온']]
    low = low[['일', '최저기온']]

    # 그룹핑 하기
    temp_g = round(temp.groupby(temp['일'], as_index=False).mean(), 4)
    preci_g = preci.groupby(preci['일'], as_index=False).sum()
    hum_g = round(hum.groupby(hum['일'], as_index=False).mean(), 4)
    high_g = high.groupby(high['일'], as_index=False).max()
    low_g = low.groupby(low['일'], as_index=False).min()

    # 병합하기
    df_1 = pd.merge(temp_g, preci_g)
    df_2 = pd.merge(df_1, hum_g)
    df_3 = pd.merge(df_2, high_g)
    df = pd.merge(df_3, low_g)

    # 해충지수 구하기
    watersides = [round(i * 5.552 + 12.157, 3) for i in df['최저기온']]
    residences = [round((i*3.346) - (j*1.114)+0.273, 3)
              for i, j in zip(df['평균기온'], df['최고기온'])]
    parks = [round((i*3.290) + (j*0.442) - (k*0.082) - 21.349, 3)
         for i, j, k in zip(df['평균기온'], df['최고기온'], df['강수량'])]

    # 해충지수 열 추가하기
    df['수변부'] = watersides
    df['주거지부'] = residences
    df['공원부'] = parks
    df['동이름'] = name

    # 저장하기
    df.to_csv(f'../../../result/8월/8월_{name}_모기지수.csv', encoding='utf-8', index=False)

    # ---- 여기까지 쓰면 됨.
```
